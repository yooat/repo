```python
y = df['Price']                        # fix the "y" var as one to predict
plt.figure(figsize=(30,30))

for idx, col in enumerate(df.columns): # return each idx, col of df
    try:

        if type(df[col][0]) != str:    # if data type of col is not str, return subplot

            x = df[col]
            plt.subplot(7,6,idx); plt.scatter(x,y); plt.title(col)
        else:                         # else data type of col is str, return blank subplot

            plt.subplot(7,6,idx); plt.scatter([],[]); plt.title(col)
    
    except:
        continue # when a error occur, ignore
        
plt.show()
```
